```python
l_df=[]
for i in da.select_dtypes(include=object):
    cf=da[i].value_counts()
    keys=cf.keys()
    values=cf.values
    cols=["labels","count"]
    df=pd.DataFrame(zip(keys,values),columns=cols)
    l_df.append(df)
    df.to_csv(f"{i}.csv",index=False)
```

**bar char using seaborn**
- iin matplotlib the method is "plt.bar"
- in seaborn the method is "countplot"
- in matplot,
    - we wil take the raw data
    -  we will create our frequency table i.e., dataframe
    - we will pass the column 1,column2 and dataframename

- in seaborn countplot
    - it takes direct raw data
    - we will pprovide direct column name
    - we will provide main data frame name
