```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

**reading the data set**

```python
bank=pd.read_csv(r"C:\Users\HP\EDA\dataset\bank.csv",sep=";")

```

**analysing the data using seaborn**

**creating  bank_graphs_seaborn directory**

```python
try:
    root_directory=os.getcwd()
    new_folder="bank_graphs_seaborn"
    new_dir=os.path.join(root_directory,new_folder)
    os.makedirs(new_dir)
except Exception as e:
    print(e)
```

```python
for i in bank.select_dtypes(include="object").columns:
    plt.figure(figsize=(15,5))
    sns.countplot(data=bank,x=i,order=bank[i].value_counts().keys())
    plt.title(i.capitalize())
    plt.xlabel("LABEL")
    plt.ylabel("COUNT")
    plt.savefig(f"{new_dir}\\{i}.jpg")
    plt.show()
```

**Analysing the data using matplotlib**

**creating the "value_counts_df" to store all the frequency tables**

```python
import os
try:
    new_dir=os.path.join(os.getcwd(),"value_counts_df")
    os.makedirs(new_dir)

except Exception as e:
    print(e)
```

```python

for i in bank.select_dtypes(include="object").columns:
    b=bank[i].value_counts()
    k=b.keys()
    v=b.values
    cols=["labels","count"]
    bank_df=pd.DataFrame(zip(k,v),columns=cols)
    bank_df.to_csv(f"{new_dir}\\{i}.csv",index=False)
    
```

**creating "matplotlib_graphs"  to store all the graphs**

```python
try:
    dir=os.path.join(os.getcwd(),"matplotlib_graphs")
    os.makedirs(dir)
except Exception as e:
    print(e)
    
```

```python
dfs=os.listdir(new_dir)
for i in dfs:
    df=pd.read_csv(f"{new_dir}\\{i}")
    #plt.figure(figsize=(10,5))
    print(df)
    plt.bar("labels","count",data=df)
    plt.xlabel("LABEL")
    plt.ylabel("COUNT")
    plt.title(i)
    plt.savefig(f"{dir}\\{i}.jpg")
    plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
bank=pd.read_csv(r"C:\Users\HP\EDA\dataset\bank.csv",sep=";")

cat_cols=bank.select_dtypes(include="object").keys()
cat_cols
```

```python
for i,j in zip(cat_cols,range(1,12)):

    ax=bank[i].value_counts()
    ax_graph=ax.plot(kind="bar")
    ax_graph.bar_label(ax_graph.containers[0])
    plt.show()
```

```python
for i,j in zip(cat_cols,range(1,len(cat_cols)+1)):
    keys=bank[i].value_counts().keys()
    values=bank[i].value_counts().values
    plt.pie(x=values,explode=[0.1]*len(keys),labels=keys,radius=1.5,startangle=180,autopct="%0.2f%%")
    plt.show()
```
