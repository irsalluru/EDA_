- Whenever you want to create any list from series
- first convert theseries into a dictionary

```python
l1=[]
l2=[]
data_types=visa_df.dtypes
dict(data_types)
for i,j in data_types.items():
    if j== 'object':
        l1.append(i)

    else:
        l2.append(i)
l1,l2
```

```python
cat_list=[i for i,j in data_types.items() if j=="object"]
num_list=[i for i,j in data_types.items() if j!="object"]
cat_list,num_list
```

```python
len(cat_list),len(num_list)
```

- we cannot create a dataframe using these two lists

```python
d1=pd.DataFrame(cat_list,columns=["CAT"])
d1.to_csv('categoricaldata.csv',index=False)
d2=pd.DataFrame(num_list,columns=["NUM"])
d2.to_csv('numericaldata.csv',index=False)
```

**take**
- first argument is indices
     - indices will take some values in a list
     - it tries to extract the data
     - the values we provide represents coloumns or rows?
- second argument axis
    - axis=0 ----- rows
    - axis=1 ------ coloumns
    - by default it is zero 

```python
visa_df.take([1,2,3])
```

```python
visa_df.take([1,2,3],axis=1)
```

```python
visa_df.take([i for i in visa_df.continent if(visa_df.continent=="Asia") ],axis=0)
```

```python
visa_df.take([2,5,7],axis=1)
```

```python
visa_df.take([2:7])
```
