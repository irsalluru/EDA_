```python
import pandas as pd
d1=pd.DataFrame() # creating a data frame
d1
```

```python
l1=['ram','ravi','sita']
d1=pd.DataFrame(l1)  # inserting data into  a data frame
d1

```

```python
l2=[20,25,20]
d1=pd.DataFrame(zip(l1,l2)) # inserting two lists at once
d1
```

```python
d1=pd.DataFrame(zip(l1,l2),columns=['name','age']) # Giving names to the coloums
d1
```

```python
d1['income']=[20000,15000,30000] # Adding a new coloum to the dataframe nate that the length should be same as remaining coloums
d1

```

```python
d1.columns # returns the all cloumn names
```

```python
#droping  coloumn here we have to consider two variable "axis" and"inplace"
#axis is used to represent coloums or rows by default it is 0 (i.e., rows) and for coloums we have to assign 1 value
#inplace is a boolean variable, if true the changes will be applied to original data frame, 
#if false(by default) the changes will be done but we have to store in different dataframe
d1.drop('income',axis=1,inplace=True) 
d1 
```

```python
d2=d1.drop('age',axis=1) 
d2

```
