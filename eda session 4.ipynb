```python

#outliners in the given dataset

Q1=round(np.percentile(visa_df["prevailing_wage"],25))
Q2=round(np.percentile(visa_df["prevailing_wage"],50))
Q3=round(np.percentile(visa_df["prevailing_wage"],75))
IQR=Q3-Q1
LB=Q1-1.5*IQR
UB=Q3+1.5*IQR
con1=visa_df["prevailing_wage"]<LB
con2=visa_df["prevailing_wage"]>UB
con3=con1 | con2
outliners=visa_df[con3]
outliners[]
```

```python

#NON outliners in the given dataset

Q1=round(np.percentile(visa_df["prevailing_wage"],25))
Q2=round(np.percentile(visa_df["prevailing_wage"],50))
Q3=round(np.percentile(visa_df["prevailing_wage"],75))
IQR=Q3-Q1
LB=Q1-1.5*IQR
UB=Q3+1.5*IQR
con1=visa_df["prevailing_wage"]>LB
con2=visa_df["prevailing_wage"]<UB
con3=con1 & con2
nonoutliners=visa_df[con3]
nonoutliners
```

**how to deal outliers**
1. we can drop the outliers
     - droping ourliers is an easy task, but not recommended
   - if you drop the outliers we are dropping the rows
   - which means we are deroping the other column values also
   - may be some some important info in the other columns will be lost
   - 

2. Fill with median
   - we already know that outliners doesnot affect median
   - so it is the best practice to fill outliners with medain
     
4. Capping or flooring
   - instead of fill with median
   - we can fill with certain vlues
   - the outliners which are less than lB fill with LB value
   - th eoutliers which are greater than UB fill with UB value

- In numpy we have method called where
- it has 3 arguments
      - condition
      - value if condition is true
      - value if condition is false
- it is applicable for binary means if and else

```python
list1=[]
wage_data=visa_df["prevailing_wage"]
for value in wage_data:
    if value<LB or value >UB:
        median=wage_data.median()
        list1.append(median)
    else:
        list1.append(value)

```

```python
dict1={"Sub":['DS','DE','CS'],
         'marks':[100,200,300]}
df=pd.DataFrame(dict1)
df
```
